# Make toy example 


    3 cell types
    1 level of binary batch
    

```r
fig.size <- function (height, width) {
    options(repr.plot.height = height, repr.plot.width = width)
}
```

```r
model.matrix.full <- function(vars_use, data_df, intercept = 1) {
    contrasts_df <- lapply(vars_use, function(x) {
        diag(nlevels(data.frame(data_df)[, x, drop = TRUE]))
    })
    names(contrasts_df) <- vars_use

    model.matrix(as.formula(sprintf("~ %d + %s", intercept, paste(vars_use, collapse = "+"))), 
                 data=data_df, contrasts.arg=contrasts_df)
    
}
```

## Single Batch

```r
N <- 100
K <- 4
B <- 2
mu <- 1:K

meta_data <- data.table(cell_id = paste("X", 1:N), batch = factor(sample(B, N, TRUE)), cell_type = factor(sample(LETTERS[1:K], N, TRUE)))
# phi <- model.matrix.full("batch", meta_data) %>% t
phi <- model.matrix(~ 1 + batch, meta_data) %>% t

W <- matrix(c(mu, rnorm((nrow(phi) - 1) * K, 0, .2)), K, nrow(phi)) %>% t

## deterministic cluster assignment 
R <- t(model.matrix.full("cell_type", meta_data, intercept = 0))

# dim(W); dim(phi)
```

```r
with(meta_data, table(batch, cell_type))
```

### Generate data

$Z_i \leftarrow \sum_k R_{ki} \cdot W^T\phi + \epsilon_k $

$\epsilon_k \sim \mathcal{N}(0, \sigma^2_k)$

```r
Z <- colSums(R * crossprod(W, phi)) + rnorm(N, 0, .05)
```

```r
fig.size(3, 6)
cbind(meta_data, Z = Z) %>% 
    ggplot(aes(cell_type, Z, col = batch)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21)
```

### Estimate $w_t$

```r
library(matlib)

W_est <- Reduce(cbind, lapply(1:K, function(i) {
    inv(phi %*% diag(R[i, ]) %*% t(phi)) %*% phi %*% diag(R[i, ]) %*% Z
}))

    
```

### Correct $z$

```r
Z_corr <- Z - colSums(R * crossprod(W_est[2:nrow(W_est), , drop = FALSE], phi[2:nrow(phi), , drop = FALSE]))
```

```r
library(patchwork)
fig.size(3, 8)
cbind(meta_data, Z = Z) %>% 
    ggplot(aes(cell_type, Z, col = batch)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) +
        labs(title = "Before correction") + guides(col = FALSE) + 

cbind(meta_data, Z = Z_corr) %>% 
    ggplot(aes(cell_type, Z, col = batch)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) + 
        labs(title = "After correction") + guides(col = FALSE)
```

## Two batches

```r
N <- 100
K <- 4
B1 <- 2
B2 <- 3
mu <- 1:K

meta_data <- data.table(cell_id = paste("X", 1:N), 
                        batch2 = factor(sample(B1, N, TRUE)), 
                        batch1 = factor(sample(B2, N, TRUE)), 
                        cell_type = factor(sample(LETTERS[1:K], N, TRUE)))
# phi <- model.matrix.full("batch", meta_data) %>% t
phi <- model.matrix(~ 1 + batch1 + batch2, meta_data) %>% t

W <- matrix(c(mu, rnorm((nrow(phi) - 1) * K, 0, .2)), K, nrow(phi)) %>% t

## deterministic cluster assignment 
R <- t(model.matrix.full("cell_type", meta_data, intercept = 0))

# dim(W); dim(phi)
```

```r
with(meta_data, table(batch1, batch2, cell_type))
```

### Generate data

$Z_i \leftarrow \sum_k R_{ki} \cdot W^T\phi + \epsilon_k $

$\epsilon_k \sim \mathcal{N}(0, \sigma^2_k)$

```r
Z <- colSums(R * crossprod(W, phi)) + rnorm(N, 0, .05)
```

```r
fig.size(3, 10) 
cbind(meta_data, Z = Z) %>% 
    ggplot(aes(cell_type, Z, col = batch1)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) + 
        guides(color = FALSE) + 

cbind(meta_data, Z = Z) %>% 
    ggplot(aes(cell_type, Z, col = batch2)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) +
        guides(color = FALSE)
```

### Estimate $w_t$

```r
library(matlib)

W_est <- Reduce(cbind, lapply(1:K, function(i) {
    inv(phi %*% diag(R[i, ]) %*% t(phi)) %*% phi %*% diag(R[i, ]) %*% Z
}))

    
```

### Correct $z$

```r
Z_corr <- Z - colSums(R * crossprod(W_est[2:nrow(W_est), , drop = FALSE], phi[2:nrow(phi), , drop = FALSE]))
```

```r
library(patchwork)
fig.size(6, 8)
cbind(meta_data, Z = Z) %>% 
    ggplot(aes(cell_type, Z, col = batch1)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) +
        labs(title = "Before correction") + guides(col = FALSE) + 

cbind(meta_data, Z = Z_corr) %>% 
    ggplot(aes(cell_type, Z, col = batch1)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) + 
        labs(title = "After correction") + guides(col = FALSE) + 

cbind(meta_data, Z = Z) %>% 
    ggplot(aes(cell_type, Z, col = batch2)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) +
        labs(title = "Before correction") + guides(col = FALSE) + 

cbind(meta_data, Z = Z_corr) %>% 
    ggplot(aes(cell_type, Z, col = batch2)) + 
        geom_point(position = position_jitter(height = 0, width = .3), shape = 21) + 
        labs(title = "After correction") + guides(col = FALSE)
```

# Function

    Given: R, Z, and Phi
    Return: Z_corr

```r
library(matlib)

MOEcorrect <- function(Z, R, phi) {
    W <- Reduce(cbind, lapply(1:K, function(i) {
        inv(phi %*% diag(R[i, ]) %*% t(phi)) %*% phi %*% diag(R[i, ]) %*% Z
    }))

    Z - colSums(R * crossprod(W[2:nrow(W), , drop = FALSE], phi[2:nrow(phi), , drop = FALSE]))    
}

```

```r
batch_labels <- pbmc@meta.data$stim
```

```r


## RUN HARMONY
harmonyObj <- new(harmony, 0) ## 0 is a dummy variable - will change later
batch_mat <- t(onehot(batch_labels))
batch_mat_moe <- model.matrix(~ 1 + stim, pbmc@meta.data) %>% t

harmonyObj$setup(
    pc_mat, ## Z
    batch_mat, ## Phi
    sigma, ## sigma
    theta, ## theta
    max.iter.cluster, ## max.iter
    epsilon.cluster, ## kmeans converge.thresh
    epsilon.harmony, ## harmony epsilon
    TRUE, ## correct Z_orig only
    alpha, ## EXPERIMENTAL: alpha, strength of dirichlet prior for DKL penalty
    nclust, ## K
    tau, ## tau (desired cells/cluster)
    block.size, ## model$block.size
    rep(1, length(batch_labels)), ## EXPERIMENTAL FEATURE: each cell gets its own weight
    FALSE, ## do linear correction on Z_cos?
    rep(1, nrow(batch_mat)), ## EXPERIMENTAL FEATURE: only correct certain batches
    burn.in.time ## window size for kmeans convergence
)
```
